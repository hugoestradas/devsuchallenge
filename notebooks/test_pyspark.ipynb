```python
# Ejercicio # 1  
conn = sqlite3.connect('test.db')


####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
                        WITH AnalisisResultados AS (
                            SELECT 
                                trim(substr(partido, 1, instr(partido, ' vs. ') - 1)) AS Equipo1,
                                trim(substr(partido, instr(partido, ' vs. ') + 4)) AS Equipo2,
                                CAST(substr(resultado, 1, instr(resultado, ',') - 1) AS INT) AS Marcador1,
                                CAST(substr(resultado, instr(resultado, ',') + 1) AS INT) AS Marcador2,
                                grupo AS Grupo
                            FROM Partidos
                        ),
                        PuntosGoles AS (
                            SELECT 
                                Equipo1 AS Equipo,
                                Grupo,
                                CASE 
                                    WHEN Marcador1 > Marcador2 THEN 3
                                    WHEN Marcador1 = Marcador2 THEN 1
                                    ELSE 0
                                END AS Puntos,
                                Marcador1 AS GolesAnotados,
                                Marcador2 AS GolesConcedidos
                            FROM AnalisisResultados
                            UNION ALL
                            SELECT 
                                Equipo2 AS Equipo,
                                Grupo,
                                CASE 
                                    WHEN Marcador2 > Marcador1 THEN 3
                                    WHEN Marcador1 = Marcador2 THEN 1
                                    ELSE 0
                                END AS Puntos,
                                Marcador2 AS GolesAnotados,
                                Marcador1 AS GolesConcedidos
                            FROM AnalisisResultados
                        )
                        --SELECT * FROM PuntosGoles;
                        SELECT 
                            Equipo,
                            Grupo,
                            SUM(Puntos) AS Puntos,
                            SUM(GolesAnotados) - SUM(GolesConcedidos) AS Diferencia
                        FROM PuntosGoles
                        GROUP BY Equipo, Grupo
                        ORDER BY Grupo, Puntos DESC, Diferencia DESC;
                       ''') 

for row in cursor:
  print(row)
conn.close()
```

```python
# Ejercicio # 2
conn = sqlite3.connect('test.db')

####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
                       WITH ClasificadosGrupo AS (
                            SELECT 
                                equipo,
                                grupo,
                                puntos,
                                gol_diferencia,
                                ROW_NUMBER() OVER (
                                    PARTITION BY grupo 
                                    ORDER BY puntos DESC, gol_diferencia DESC
                                ) AS rank
                            FROM Resultados_Qatar
                        )
                        SELECT 
                            cg.equipo,
                            cg.grupo,
                            cg.puntos,
                            cg.gol_diferencia,
                            SUBSTR(c.sede, INSTR(c.sede, ', ') + 2) || ', ' || c.fecha || ' ' || c.hora as 'Ciudad/Fecha'
                        FROM ClasificadosGrupo cg
                        JOIN Clasificados c
                            ON c.partido LIKE '%' || 
                                CASE 
                                    WHEN cg.rank = 1 THEN '1º del ' || cg.grupo
                                    WHEN cg.rank = 2 THEN '2º del ' || cg.grupo
                                END || '%'
                        WHERE cg.rank <= 2
                        ORDER BY cg.grupo, cg.rank;
                       ''')

for row in cursor:
  print(row)
conn.close()
```
